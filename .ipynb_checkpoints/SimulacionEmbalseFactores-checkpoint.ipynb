```python
import os
import numpy as np
import pandas as pd
import datetime as dt
import math

from matplotlib.pylab import *
from mpl_toolkits.axes_grid1 import host_subplot
```

```python
def simulacionembalse(path, name1, name2, name3, name4, hembalse , fmuerto, pf, factorinfiltracion, factorevapora, values):

  # name1 : Demanda
  # name2 : altura_volumen_area
  # name3 : mmf model parámetro
  # name4 : dataset hidrologia

  # Read Documents
    from datetime import datetime
    dateparse = lambda x: datetime.strptime(x, '%m/%d/%Y')

    train_path1 = os.path.join(path,f'{name1}.csv')
    train_path2 = os.path.join(path,f'{name2}.csv')
    train_path3 = os.path.join(path,f'{name3}.csv')
    train_path4 = os.path.join(path,f'{name4}.csv')
 
    demanda      = pd.read_csv(train_path1)
    hvolarea     = pd.read_csv(train_path2)
    coeficientes = pd.read_csv(train_path3)
    hidrometdat  = pd.read_csv(train_path4, parse_dates=['date'], date_parser=dateparse)

  # Demanda
    DemandaTotal          = demanda.iloc[:,-1].sum()
    demanda['percentage'] = demanda.iloc[:,-1]/DemandaTotal
    percenmon             = list(np.array(demanda.percentage))

  # Volumenes
  ## Coeficientes
    mmf = np.array(coeficientes.iloc[:,-1])

  ## Volumen Embalse
    VolEmbalse  = hvolarea.volumen[hvolarea.altura == hembalse].values[0]
    CotaEmbalse = hvolarea.cota[hvolarea.altura == hembalse].values[0]
    AreaEmbalse = hvolarea.area[hvolarea.altura == hembalse].values[0]

  ## Volumen Muerto
    hmuerto     = fmuerto[0]
    CotaMuerto  = hvolarea.cota[hvolarea.altura == fmuerto[0]].values[0]
    VolMuerto   = hvolarea.volumen[hvolarea.altura == fmuerto[0]].values[0]
    AreaMuerto  = hvolarea.area[hvolarea.altura == fmuerto[0]].values[0]

  # Condiciones iniciales embalse lleno
    VolInicial  = VolEmbalse
    AreaInicial = AreaEmbalse

  # Días de cada mes simulado : days
    from datetime import datetime
    dateparse = lambda x: datetime.strptime(x, '%d/%m/%Y')
    
    days = np.array(hidrometdat.iloc[:,0].dt.daysinmonth, dtype=float)

  # Volumen aporte
    volaporte     = (np.array(hidrometdat.iloc[:,1])*86400*days)  # en m3/s
  # Precipitacion
    precipitation = np.array(hidrometdat.iloc[:,2])

  # Demanda simulada por mes (por los % de demanda) : voldemanda
    numtimes          = int((len(volaporte)/12))
    percentagedemanda = np.array(percenmon*(numtimes+1))[:len(precipitation)]

  # Evaporacion promedio mes(mm) * factor evaporacion
    evaporacion = np.array(hidrometdat.iloc[:,3])*days*factorevapora

  # Array zeros

    areafvol         = np.zeros_like(days)
    volprecipitation = np.zeros_like(days)
    infiltrationloss = np.zeros_like(days)
    volevaporacion   = np.zeros_like(days)
    volfinalmes      = np.zeros_like(days)
    estado           = np.zeros_like(days)

    pf5  = np.array([])
    pf10 = np.array([])
    pf15 = np.array([])

    pf5_r  = np.array([])
    pf10_r = np.array([])
    pf15_r = np.array([])    
    
    factors = np.arange(values[0], values[1], values[2])   # values[] : probabilidades de falla
    for factor in factors:
        
        volefectivo   = DemandaTotal*factor # voldemanda(input)
        voldemandames = volefectivo*percentagedemanda
        
        
        for i in range(0,len(days)):
            
            if i == 0:
                areafvol[i] = AreaInicial
                volprecipitation[i] = (precipitation[i]/1000)*areafvol[i]
                infiltrationloss[i] = factorinfiltracion*areafvol[i]*days[i]/1000
                volevaporacion[i]   = areafvol[i]*(evaporacion[i]/1000)+infiltrationloss[i]
                volfinalmes[i]      = VolInicial + volprecipitation[i] + volaporte[i] - (voldemandames[i]+volevaporacion[i])
                
                #print(areafvol[i], volprecipitation[i],infiltrationloss[i],volevaporacion[i],voldemandames[i], volfinalmes[i])
            else:
                if (volfinalmes[i-1] > VolMuerto and volfinalmes[i-1] < VolEmbalse):
                    areafvol[i] = (mmf[0]*mmf[1]+mmf[2]*(volfinalmes[i-1]**mmf[3]))/(mmf[1]+volfinalmes[i-1]**mmf[3])
                elif volfinalmes[i-1] > VolEmbalse:
                    areafvol[i] = AreaEmbalse
                else:
                    areafvol[i] = AreaMuerto

                volprecipitation[i] = (precipitation[i]/1000)*areafvol[i]
                infiltrationloss[i] = factorinfiltracion*areafvol[i]*days[i]/1000
                #print(areafvol[i])
                volevaporacion[i]   = areafvol[i]*(evaporacion[i]/1000)+infiltrationloss[i]
                inoutvol = volprecipitation[i] + volaporte[i] - (voldemandames[i]+volevaporacion[i])

                if (volfinalmes[i-1] > VolMuerto and volfinalmes[i-1] < VolEmbalse):
                    volfinalmes[i] = volfinalmes[i-1] + inoutvol
                elif volfinalmes[i-1] > VolEmbalse:
                    volfinalmes[i] = VolEmbalse + inoutvol
                else:
                    volfinalmes[i] = VolMuerto + inoutvol
            
                #print(areafvol[i],volprecipitation[i],volaporte[i],voldemandames[i],areafvol[i],infiltrationloss[i],volevaporacion[i], volfinalmes[i])
            
            if (volfinalmes[i] > VolMuerto and volfinalmes[i] < VolEmbalse):
                estado[i] =  1 # OK
            if volfinalmes[i] > VolEmbalse:
                estado[i] =  2 # REBALSA
            if volfinalmes[i] < VolMuerto:
                estado[i] =  3 # Menor a volumen muerto

        # NRO de Fallas
        nropf = len(pf)

        p5  = len(days)*pf[0]/100 # numero de meses simulados para cumplir 5%, deber ser un integer

        p10 = len(days)*pf[1]/100
        p15 = len(days)*pf[2]/100
        
        

        nrofallas5  = p5  if (p5  - int(p5)  == 0) else math.ceil(p5)
        #print(p5, math.ceil(p5), nrofallas5)
        nrofallas10 = p10 if (p10 - int(p10) == 0) else math.ceil(p10)
        nrofallas15 = p15 if (p15 - int(p15) == 0) else math.ceil(p15)
        

        nrofallas  = np.count_nonzero(estado == 3)
        nrofallas2 = len(np.where(estado == 3)[0])*100/len(days)
        

        if nrofallas == nrofallas5:
            pf5 = np.append(pf5, factor)
        if len(pf5) == 0 and nrofallas<nrofallas5+1:
            pf5_r = np.append(pf5_r, factor)
                
        if nrofallas == nrofallas10:
            pf10 = np.append(pf10, factor)
        if len(pf10) == 0 and nrofallas<nrofallas10+1:
            pf10_r = np.append(pf10_r, factor)          
            
        if nrofallas == nrofallas15:
            pf15 = np.append(pf15, factor)
        if len(pf15) == 0 and nrofallas<nrofallas15+1:
            pf15_r = np.append(pf15_r, factor)
            
        if nrofallas > nrofallas15:
            break
            
    if len(pf5)>0:
        pf5 = pf5
    elif len(pf5)==0:
        pf5 = pf5_r
        
    if len(pf10)>0:
        p10 = p10
    elif len(pf10)==0:
        pf10 = pf10_r        
        
    if len(pf15)>0:
        pf15 = pf15
    elif len(pf15)==0:
        pf15 = pf15_r  
    print(len(pf5_r),len(pf10_r),len(pf15_r))
            
    df_fallas = pd.DataFrame(data=estado, index= pd.to_datetime(hidrometdat.iloc[:,0]), columns=['estado'])
    
    df_fallas_freq = pd.DataFrame(columns=['OK','REBALSA','FALLA'])
    
    for i in range(1,13):
        df_fallas_mes = df_fallas[df_fallas.index.month==i]
        nro_ok = len(np.where(df_fallas_mes.iloc[:,-1] == 1)[0])
        nro_rebalse = len(np.where(df_fallas_mes.iloc[:,-1] == 2)[0])
        nro_falla   = len(np.where(df_fallas_mes.iloc[:,-1] == 3)[0])
        
        df_fallas_freq.loc[f'{i}'] = [nro_ok, nro_rebalse, nro_falla]
        
    df_fallas_freq.index.name = 'mes'
    df_fallas_freq.to_csv(f'frecuencia_estado_h{hembalse}.csv')
    
    #print(nrofallas5,nrofallas10,nrofallas15) # Nro de fallas, redondeo al inmediato superior si pf contiene decimales
    print(nrofallas5, nrofallas10, nrofallas15)
    print(f"Altura embalse : ", hembalse, "| Altura embalse muerto : ",hmuerto)# nrofallas, factor)
    print(f"Nro soluciones", "pf5:",len(pf5),"| pf10:", len(pf10),"| pf15:" , len(pf15))
    print('Found it')
    
    
    factores = np.array([pf5[-1],pf10[-1],pf15[-1]])
    
    return  factores
```

```python
def SimulacionFactoresDF(path, name1, name2, name3, name4, hembalse , fmuerto, pf, factorinfiltracion, factorevapora, values, nameout, outputpath):
    

    probfalla = pd.DataFrame(columns=[f'{pf[0]}%',f'{pf[1]}%',f'{pf[2]}%'])

    for h in hembalse:

        probfalla.loc[h] = simulacionembalse(path, name1, name2, name3, name4, h , fmuerto, 
                                             pf, factorinfiltracion, factorevapora, values)

    probfalla.index.name = 'altura'




    probfalla.to_csv(f'{nameout}.csv')

    return probfalla
```
